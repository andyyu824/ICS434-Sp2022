#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
disciplines = ["CS", "Engineering", "Math", "Physics" ]

np.random.seed(42)
arr = np.random.choice(disciplines, 1000)
students_affiliations = np.unique(arr, return_counts=True)

students_affiliations = list(zip(students_affiliations[0], students_affiliations[1]))
students_affiliations = sorted(students_affiliations, key = lambda x: x[1], reverse=True)

students_affiliations = list(zip(*students_affiliations))
```

```python
plt.figure(figsize=(12,8))

plt.bar(students_affiliations[0], students_affiliations[1], color='#FF5733', alpha=0.55)

plt.xlabel("Disciplines")
plt.ylabel("Counts")

title = plt.title("Distribution of Students in the Program", fontsize=20)
plt.setp(title, color='g') 
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
np.random.seed(42)

plt.figure(figsize=(12,6))

random_x = np.random.randint(range(0,200), 1000)
random_y = np.random.randint(range(0,200), 1000)

plt.scatter(random_x, random_y, marker='s')


plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.title("Random 42")
```
